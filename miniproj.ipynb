Using Neural Networks and Importing the necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Loading the datasets
data=sklearn.datasets.load_breast_cancer()
```

```python
#Loading data into Pandas Data Frame
data_frame=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
data_frame.head() #to display the first 5 values present in the dataset
```

```python
data_frame.shape   #to get the shape of dataset in form of number of rows and columns
```

```python
data_frame.isnull().sum()     #to get output of number of empty/null variables
```

```python
data_frame['Label']=data.target
```

```python
data_frame.head()
```

```python
data_frame['Label'].value_counts()
```

```python
X=data_frame.drop('Label',axis=1)
Y=data_frame['Label']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
#To standardize the data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train_fit=scale.fit_transform(X_train)
X_test_fit=scale.transform(X_test)
```

BUILDING NEURAL NETWORKS

```python
#importing tensorflow and keras library
#TensorFlow-deeplearning library by Google(Alternate-pytouch)
#Keras-not a standalone library but a wrapper for TensorFlow
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

Setting up Layers for Neural Network

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)), #input
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(15,activation='relu'),
                        keras.layers.Dense(10,activation='relu'),#hidden-20 neurons in it relu-rectify linear unit
                        keras.layers.Dense(2,activation='sigmoid')  #output-2 neurons in it
]) #Stacking up the input,hidden and output layers
#Dense-all neurons in that layer are connected to previous layer
#flatten-to convert into 1-D array one neuron for each feature
#number of neurons in output layer=number of cases or ouptut cases in this case 0&1 so 2 neurons
```

Compiling Neural Networks

```python
model.compile(optimizer='adam',              #to optimize this network using adam function
              loss='sparse_categorical_crossentropy',     #using this loss function as only two output variables are available that is 0&1
              metrics=['accuracy'])
```

Training to obtain accuracy of 99.27%

```python
#Training Neural Network
history=model.fit(X_train_fit,Y_train,validation_split=0.1,epochs=20) #validation-divide to train and test data simultaneously
```

Accuracy of the Test Data

```python
loss,accuracy=model.evaluate(X_test_fit,Y_test)
print(accuracy)
```

```python
Y_pred=model.predict(X_test_fit)
print(Y_pred.shape)
```

Predict Function gives the probability of whether the label is 1 or 0 but does give the value
